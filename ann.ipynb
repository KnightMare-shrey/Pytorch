```python
import torch
import torch.nn as nn
```

```python
class ANN(nn.Module):
    def __init__(self,input_layers,output_layers):
        super(ANN,self).__init__()
        self.input_layer = nn.Linear(input_layers,784)
        self.hidden_layer1=nn.Sequential(
            nn.Linear(784,256),
            nn.ReLU(),
            nn.Linear(256,128),
            nn.ReLU(),
            nn.Linear(128,64),
            nn.ReLU()
        )
        self.output_layer =nn.Sequential(
            nn.Linear(64,32),
            nn.ReLU(),
            nn.Linear(32,output_layers)
        )

    def forward(self,x):
        x=self.input_layer(x)
        x=self.hidden_layer1(x)
        x=self.output_layer(x)

        return x
```

```python
# test ANN
x=torch.randn(32,1,784)
model=ANN(784,10)
y=model(x)
print(y.shape)
print(y)
```
